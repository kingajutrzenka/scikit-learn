```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.figure_factory as ff
```

```python
sns.set()
```

```python
np.random.seed(10)
np.set_printoptions(precision=6, suppress=True)

def sigmoid(x):
  return 1 / (1 + np.exp(-x)) #funkcja sigmoid

x = np.arange(-5, 5, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.title('Sigmoid function')
plt.show()

# funkcja sigmoid spaszcza wartoci do 0-1
```

# Dane

```python
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
print(data)
```

```python
x = data.data
y = data.target

print(x.shape) #569 danych
print(y.shape)
```

```python
df = pd.DataFrame(data = x, columns=data.feature_names)
df.info()
```

```python
df.describe().T
```

```python
df.head()
```

# Podzia danych

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Dopasowanie modelu

```python
from sklearn.linear_model import LogisticRegression
classifien = LogisticRegression()
classifien.fit(x_train, y_train) #trenujemy
```

# Predykcja na podstawie modelu

```python
y_prob = classifien.predict_proba(x_test)
print(y_prob) # w wierszach sumuje si do 1
```

```python
y_pred = classifien.predict(x_test)
print(y_pred)
# prawdopodobiestwo kl. 0 > 1 daje 0, kl. 1 > 0 daje 1
```

# Ocena modelu

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from mlxtend.plotting import plot_confusion_matrix

cm = confusion_matrix(y_test, y_pred)
plot_confusion_matrix(cm)

acc = accuracy_score(y_test, y_pred)
print(acc)

# staa funkcja ale nieinteraktywny :/
```

```python
def plot_confusion_matrix(cm):
  cm = cm[::-1]
  cm = pd.DataFrame(cm, columns=['pred_0', 'pred_1'], index=['true_1', 'true_0'])

  fig = ff.create_annotated_heatmap(
      z=cm.values,
      x=list(cm.columns),
      y=list(cm.index),
      colorscale = 'ice',
      showscale=True,
      reversescale=True
  )

  fig.update_layout(width=400, height=400, title=f'Confusion Matrix, Accuracy: {acc:.4f}', font_size=16)
  fig.show()

plot_confusion_matrix(cm)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred, target_names=data.target_names))
```
