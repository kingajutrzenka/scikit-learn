# Przykad

```python
import numpy as np

x = np.arange(0, 50, 0.5)
noise = 10 * np.random.randn(100)
y = 2 * x + 100 + noise
x = x.reshape(-1, 1)

print('Rozmiar X:',x.shape)
print('Rozmiar y:',y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)

print('Rozmiar X_train:',x_train.shape)
print('Rozmiar y_train:',y_train.shape)
print('Rozmiar X_test:',x_test.shape)
print('Rozmiar y_test:',y_test.shape)

# 75 prbek w zbiorze treningowym
# 25 w zbiorze testowym
```

```python
import seaborn as sns
sns.set()
sns.set_style('whitegrid')

plt.figure(figsize=(7, 5))
plt.scatter(x_train, y_train, color='blue', label='Dane treningowe')
plt.scatter(x_test, y_test, color='green', label='Dane testowe')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

#trenujemy na niebieskich
#ocenaimy na zielonych
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)

#trenujemy model
```

```python
print(model.coef_)
print(model.intercept_)
```

Rozwizaniem jest prosta o postaci Y = 99,15 + 2,017X

```python
# wizualizacja zbioru treningowego
plt.figure(figsize=(7, 5))
plt.title('Regresja liniowa - jedna zmienna')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x_train, y_train, color='blue', label='zbir treningowy')
plt.plot(x, model.intercept_ + model.coef_ * x, color='red')
plt.legend()
plt.show()
```

```python
# wizualizacja zbioru testowego
y_pred = model.predict(x_test)
plt.figure(figsize=(7, 5))
plt.title('Regresja liniowa - jedna zmienna')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x_test, y_test, color='green', label='zbir testowy')
plt.plot(x, model.intercept_ + model.coef_[0] * x, color='red')
plt.legend()
plt.show()
```

```python
model.score(x_test, y_test) #skuteczno modelu
```
