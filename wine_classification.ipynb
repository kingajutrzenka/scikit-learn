**Wine classification**

```python
import numpy as np
import sklearn as sk
from sklearn import datasets

np.random.seed(10)
raw_data = datasets.load_wine()
raw_data
```

```python
raw_data.keys()
```

```python
print(raw_data.DESCR)
```

```python
data = raw_data.data
target = raw_data.target
```

```python
print(data.shape)
print(target.shape) # 178 klas
```

```python
from sklearn.model_selection import train_test_split

data_train, data_test, target_train, target_test = train_test_split(data, target, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(data_train, target_train)
```

```python
target_pred = model.predict(data_test)
target_pred
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(target_test, target_pred)

""" macierz pomyłek
jest 3 x 3, bo mamy 3 klasy
- klasa 0 13 razy dobrze, 2 razy pomylona z klasą 1
- klasa 1 6 razy pomylona z klasą 0, 19 razy dobrze...
* klasa 2 zostala jako jedna idealnie sklasyfikowana
* skuteczność modelu poprawne/wszytskie * 100%"""
a = ((13+19+12)/len(target_test))*100
print(f'dokładomość testu {a}%')
```

```python
from sklearn.metrics import classification_report

print(classification_report(target_test, target_pred))
#accuracy dokładość testu
```
