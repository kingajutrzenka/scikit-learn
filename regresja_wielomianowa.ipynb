**Regresja wielomianowa**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

np.random.seed(0)
x = np.arange(-10, 10, 0.5)
noise = 80 * np.random.randn(40)
y = -x**3 + 10*x**2 - 2*x + 3 + noise # nasz wzór regresji wielorakiej
x = x.reshape(40, 1)
_ = plt.scatter(x, y)
#linia prosta nie zadziała XD
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x, y)
y_pred = regressor.predict(x)
plt.scatter(x, y)
_ = plt.plot(x, y_pred, color='red')
# ni chuja nie działa XD
```

```python
from sklearn.metrics import r2_score
r2_score(y, y_pred)

# słabeee dopasowanie, niedouczenie :/
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2) # poziom 2, bo używamy wielomianu stopnia 2
x_poly = poly_features.fit_transform(x)
x_poly
""" mamy to co napisaliśmy w podstawieniu z = [1, x, x^2]
przykładowo pierwsza kolumna to 1, druga to x, trzecia to x^2
w tym przypadku 1, -10, 100 """
```

```python
regressor_poly = LinearRegression()
regressor_poly.fit(x_poly, y)
y_pred_poly = regressor_poly.predict(x_poly)
plt.scatter(x, y)
_ = plt.plot(x, y_pred_poly, color='red')
# zdecydowanie lepsze dopasowanie
```

```python
r2_score(y, y_pred_poly)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

# model
poly_3 = PolynomialFeatures(degree=3)
x_poly = poly_3.fit_transform(x)

regressor_poly_3 = LinearRegression()
regressor_poly_3.fit(x_poly, y)
y_pred = regressor_poly_3.predict(x_poly)

# sortowanie
sort_idx = np.argsort(x[:, 0])
x_sorted = x[sort_idx]
y_pred_sorted = y_pred[sort_idx]

# wykres
plt.scatter(x, y)
plt.plot(x_sorted, y_pred_sorted, c='r')
plt.show()
# im większy stopień tym lepsze dopasowanie :)
```

```python
r2_score(y, y_pred)
```

```python
r2_score(y, y_pred)
```
